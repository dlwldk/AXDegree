#### 1. 2019년 데이터 불러오기 및 기초 통계량 확인

* 대표값(평균, 중앙값) 확인하기
* 분산 확인하기

```python
# 필요한 라이브러리 Import
import pandas as pd

# 데이터 불러오기
df_2019 = pd.read_csv('raw_data/제주특별자치도_주제3_상권분석을 위한 유동인구에 따른 카드 매출액 변화 데이터 활용_매쉬업결과_19.csv', encoding = 'cp949')
```

```python
# 데이터 확인
df_2019
```

```python
df_2019['년월'] = pd.to_datetime(df_2019['년월'])

# 연도와 월 추출
df_2019['연도'] = df_2019['년월'].dt.year
df_2019['월'] = df_2019['년월'].dt.month.astype(int)
```

```python
# 데이터 확인
df_2019.info()
```

```python
# 데이터 확인
df_2019
```

```python
# 읍면동별로 그룹화하여 월간 총 합계 계산
result = df_2019.groupby(['읍면동명','월'])

# 필요한 열 선택
result = result[['이용금액']].sum()

# 결과 출력
result
```

* 기초통계량 확인
    * Check1. 평균, 분산 확인하기
    * Check2. 범위 확인하기(Max-min)
    * Check3. 중위값과 평균 비교하기

```python
# 1. 양적변수 기초통계량 확인
## 출력 옵션 설정
pd.options.display.float_format = '{:.0f}'.format

## 데이터 프레임에서 기초 통계량 확인하기 : 평균, 중앙값, 분산
result.describe()
```

```python
# 2. 필터링으로 기초통계량 확인하기 : 건입동의 기초통계량

## 데이터 프레임에서 기초 통계량 확인하기 : 평균, 중앙값, 분산
result.loc['건입동'].describe()
```
